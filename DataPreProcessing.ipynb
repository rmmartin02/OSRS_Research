```python
def checkDates(dates):
    seen = []
    repeats = []
    for d in range(len(dates)):
        if dates[d] not in seen:
            seen.append(dates[d])
        else:
            repeats.append(d)
    return repeats
```

```python
data = {}
for item in itemPrices:
    print(item)
    dates = [a[0] for a in itemPrices[item]]
    data[item] = checkDates(dates)
```

```python
# fix duplicate dates
for item in itemPrices:
    print(item)
    dates = [a[0] for a in itemPrices[item]]
    seen = []
    repeats = []
    for d in range(len(dates)):
        if dates[d] not in seen:
            seen.append(dates[d])
        else:
            repeats.append(d)
    for index in sorted(repeats, reverse=True):
        del itemPrices[item][index]
```

```python
itemPrices = scraper.loadItemPrices()
# just throw out first 90 days of data?
for item in itemPrices:
    itemPrices[item] = itemPrices[item][90:]
```

```python
def checkOutliers(prices):
    outliers = []
    manualAdjust = []
    for i in range(len(prices)-2):
        diff1 = abs(prices[i+1]-prices[i])
        diff2 = abs(prices[i+2]-prices[i+1])
        if diff1>100 and diff2>100 and diff1>prices[i]*0.15 and diff2>prices[i+1]*0.15:
            outliers.append(i+1)
        if diff1>100 and diff2>100 and diff1>prices[i]*0.17 and diff2<prices[i+1]*0.15:
            manualAdjust.append(i+1)
    return outliers,manualAdjust
```

```python
outlier = {}
adjust = {}
for item in itemPrices:
    #print(item)
    prices = [a[1] for a in itemPrices[item]]
    data = checkOutliers(prices)
    if len(data[0])>0:
        outlier[item] = data[0]
    if len(data[1])>0:
        adjust[item] = data[1]
```
