```python
def checkDates(dates):
    seen = []
    repeats = []
    for d in range(len(dates)):
        if dates[d] not in seen:
            seen.append(dates[d])
        else:
            repeats.append(d)
    return repeats
```

```python
def checkOutliers(prices):
    outliers = []
    manualAdjust = []
    for i in range(len(prices)-2):
        diff1 = abs(prices[i+1]-prices[i])
        diff2 = abs(prices[i+2]-prices[i+1])
        if diff1>100 and diff2>100 and diff1>prices[i]*0.15 and diff2>prices[i+1]*0.15:
            outliers.append(i+1)
        if diff1>100 and diff2>100 and diff1>prices[i]*0.17 and diff2<prices[i+1]*0.15:
            manualAdjust.append(i+1)
    return outliers,manualAdjust
```

```python
outlier = {}
adjust = {}
for item in itemPrices:
    #print(item)
    prices = [a[1] for a in itemPrices[item]]
    data = checkOutliers(prices)
    if len(data[0])>0:
        outlier[item] = data[0]
    if len(data[1])>0:
        adjust[item] = data[1]
```

```python
# interpolate dates, prices, and quantities
#for item in itemPrices:
item = 'Abyssal_whip'
dates = [a[0] for a in itemPrices[item]]
total = 0
for i in range(len(dates)-1):
    diff = dates[i+1]-dates[i]
    if diff<23*hour or diff>25*hour:
        total += int(diff/day)-1
print("Before:",total)
dates = [a[0] for a in itemPrices[item]]
prices = [a[1] for a in itemPrices[item]]
quants = [a[2] for a in itemPrices[item]]
fixed = []
for i in range(len(dates)-1):
    fixed.append(itemPrices[item][i])
    diff = dates[i+1]-dates[i]
    if diff<23*hour or diff>25*hour:
        daysToAdd = int(diff/day)-1
        d = []
        p = []
        q = []
        pToAdd = (prices[i]+prices[i+1])/2
        qToAdd = (quants[i]+quants[i+1])/2
        for j in range(daysToAdd):
            d.append(dates[i]+day*(j+1))
            p.append(prices[i]+(j+1)*pToAdd)
            q.append(quants[i]+(j+1)*qToAdd)
        for j in range(daysToAdd):
            fixed.append((d[j],p[j],q[j]))
dates = [a[0] for a in fixed]
total = 0
for i in range(len(dates)-1):
    diff = dates[i+1]-dates[i]
    if diff<23*hour or diff>25*hour:
        total += int(diff/day)-1
print("After:",total)
prices = [a[1] for a in fixed]
plt.plot(prices)
```

```python
def scalePrices(x):
    diff_x = [0] * (len(x)-1)
    for i in range(len(x)-1):
        diff_x[i] = x[i+1]-x[i]
    minX = min(diff_x)
    diff_x = np.array([a+abs(minX)+.01 for a in diff_x])
    x_diff_scaled = scaler.fit_transform(diff_x.reshape(-1, 1))
    return x_diff_scaled
```

```python
# check length of available datasets
sizes = {}
for item in itemPrices:
    l = len(itemPrices[item])
    if l in sizes:
        sizes[l]+=1
    else:
        sizes[l] = 1
cumSum = 0
cumSums = []
keys = []
for key in sorted(sizes,reverse=True):
    print("%s: %s" % (key, sizes[key]))
    keys.append(key)
    cumSum += sizes[key]
    cumSums.append(cumSum)
cumSums = np.array(cumSums)/cumSum
```
